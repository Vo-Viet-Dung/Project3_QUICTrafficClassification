```python
from ctypes import sizeof
import pyshark
from pyshark.packet.packet import Packet
from scapy.all import *
import numpy as np
import csv



pcap_data = PcapReader(
    './FileTransfer/FileTransfer_00001_20180328094056.pcap')

f = open('./datasetCNN.csv', 'a')
writer = csv.writer(f)
index = 0
for packet in pcap_data:
  if index < 1000:
    # remove header of datalink packet
    payload = packet.payload
    size = len(packet.payload)
    #convert payload to byte
    payload = bytes(payload)

    row = []
    for byte in payload:
        # Nomalization data
        feature = byte/255
        row.append(feature)
    # Zeropadding
    if(size < 1400):
        for i in range(size, 1400):
            row.append(0)

    rowCSV = row
    # dataset stamp (GG Play Music: 1, File Transfer: 2)
    rowCSV.append(2)
    # for num in rowCSV:
    #     print(num)
    # print("\n")

    writer.writerow(rowCSV)
    
    index +=1
  else:
    f.close()
    break


```
