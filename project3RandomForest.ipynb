```python
# importing pandas package
import pandas as pd
  
# making data frame from csv file 
X_col = ['percent-small-packet', 'percent-medium-packet', 'percent-big-packet', 'average-payload']
data = pd.read_csv("./dataset.csv")
```

```python
train = data.sample(frac =.75)
test = data.sample(frac =.25)
train,test
```

```python
X_train = train[['percent-small-packet', 'percent-medium-packet', 'percent-big-packet', 'average-payload']]
Y_train = train[['stamp']]
X_train, Y_train
```

```python
X_test = test[['percent-small-packet', 'percent-medium-packet', 'percent-big-packet', 'average-payload']]
Y_test = test[['stamp']]
X_test, Y_test
```

```python
import numpy as np
X_train = X_train.to_numpy()
Y_train = Y_train.to_numpy()
X_train.shape, Y_train.shape
```

```python
X_test = X_test.to_numpy()
Y_test = Y_test.to_numpy()
X_test.shape, Y_test.shape
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,Y_train)

Y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred))
```
